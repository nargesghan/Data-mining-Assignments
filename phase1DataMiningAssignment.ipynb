```python
y='latitude'
fig, ax = plt.subplots(figsize=(20,15))

# Plot the specified column
ax.plot(df[y])

# Set the title
ax.set_title(y)

# Hide the top and right spines
ax.spines[['top', 'right']].set_visible(False)

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
def histogram(df, colname, num_bins=20, figscale=1):
    # Create a figure
    plt.figure(figsize=(20,15))

    # Create a histogram
    df[colname].plot(kind='hist', bins=num_bins, title=colname)

    # Hide the top and right spines
    plt.gca().spines[['top', 'right']].set_visible(False)

    # Adjust the layout
    plt.tight_layout()

    # Display the plot
    plt.show()

# Call the function with your DataFrame and column
histogram(df, 'longitude')
```

**10.Split the data into two parts: train and test. (Use 20% of the data as test data.)**

```python
# Shuffle dataframe using sample function
df = df.sample(frac=1)
df
```

```python
# Select ratio
ratio = 0.80

total_rows = df.shape[0]
train_size = int(total_rows*ratio)

# Split data into test and train
train = df[0:train_size]
test = df[train_size:]
```

```python
# print train set
print("Train dataframe")
print(train)

# print test set
print("Test dataframe")
print(test)
```
