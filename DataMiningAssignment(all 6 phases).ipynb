# **`custom transformer`**

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

# Indices for the columns
rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        # Nothing to do here, so just return self
        return self

    def transform(self, X, y=None):
        # Compute the new attributes
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]

        # Return the original data with the new attributes appended
        return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
```

```python
Custom=CombinedAttributesAdder()
data_custom_tr_top=Custom.transform(data.values)
data_custom_tr=pd.DataFrame(data_custom_tr_top)
columns=list(data)
columns.append('rooms_per_household')
columns.append('population_per_household')
columns.append('bedrooms_per_room')
data_custom_tr.columns=columns
data_custom_tr.head(10)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import FeatureUnion

# ----------------custom transform-----------------------------
rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        # Nothing to do here, so just return self
        return self

    def transform(self, X, y=None):
        # Compute the new attributes
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]

        # Return the original data with the new attributes appended
        return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]

#----------seperate categorical and numerical columns------------------

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values

#--------------drop label column------------
df=train.copy()
df_label=df['median_house_value'].copy()
df=df.drop("median_house_value",axis=1)


#---------------drop categorical data-----------------
df_num=df.drop(['ocean_proximity'],axis=1)
num_attribs=list(df_num)#سر ستون ها را برمیگرداند
cat_attribs=['ocean_proximity']

#pipeline for the numerical attributes
num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),#داده های عددی را جدا میکند
    ('imputer', SimpleImputer(missing_values=np.NaN,strategy="mean")),#مقدارهای نال را ئر میکند
    ('attribs_adder',CombinedAttributesAdder()),#اضافه کردن ستون های جدید
    ('std_scaler', StandardScaler()),#z_score
])

# pipeline for the categorical attributes
cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attribs)),#جدا کردن داده های گروهی
    ('one_hot_encoder', OneHotEncoder(sparse=False)),
])

# Combine the numerical and categorical pipelines
full_pipeline = FeatureUnion(transformer_list=[
    ("num_pipeline", num_pipeline),
    ("cat_pipeline", cat_pipeline),
])

# You can now fit and transform the data using the full pipeline
df_prepared = full_pipeline.fit_transform(df)#thise returns a numpy array

#convert numpy array to datafram:
df_prepared_df=pd.DataFrame(df_prepared,columns=['longitude',
 'latitude',
 'housing_median_age',
 'total_rooms',
 'total_bedrooms',
 'population',
 'households',
 'median_income',
 'rooms_per_house',
 'population_per_house',
 'bedrooms_per_room',
 '<1H OCEAN',
 'INLAND',
 'NEAR BAY',
 'NEAR OCEAN',
 'ISLAND'])
df_prepared_df.head()
```
